## 单层神经网络的直观理解
到目前为止我们只看到了一个神经元。如果我现在画了很多个这样的圈圈并排成一列，你能感受到我的深刻的用意（诅咒）吗？
<img src="./imgs/1l.jpeg"/>
我在干嘛？结合前一章的讲解。每个圈圈的输入一样，但是他们却可以有不同的权重，这就是说他们每一个圈都是对同一个数据做不同的理解。
比如第一个圈可以预测房子值多少钱，第二个圈可以推测适不适合做投资。第三个圈拆了后能拿多少钱等等 。现在这个W长什么样子了呢。变成了一个矩阵。每一行代表一个神经元的权重向量。
$$ W = \begin{bmatrix} 
 \color{cyan}w_{11}, \color{cyan}w_{12}, \color{cyan}w_{13} \\
 \color{cyan}w_{21}, \color{cyan}w_{22}, \color{cyan}w_{23} \\
\end{bmatrix}   \;\;\;  X= \begin{bmatrix}
 \color{red}a_1 \\
 \color{red}a_2 \\
 \color{red}a_3 \\ 
 \end{bmatrix}$$
然后感谢一下线代的优美，公式依旧是哪个公式。

$ a = g(WX + \theta) $


##  多层神经网络的直观理解
既然这个也能理解了。那么再加几列，变成下面这种剪不断理还乱的样子会怎样呢？
<img src="./imgs/nl.jpeg">

如果已经理解上面的内容。其实不论加多少列都是一个道理。每一层接受上一次的输入，处理成为结果，传递下去。这样信息随着层数的加深，含义就随之变得更加抽象

比如输入是 城市 户型这样的信息，后面一层就有了，价格，拆迁所得等概念，再往后可能就是 半夜笑醒的概率，买泰坦显卡的心痛指数,半夜老板打电话让修bug敢回一句F_ck U,倒头继续睡的勇气系数等等。

这时候公式变成什么样子了呢？ 稍微有点吓人 但很有规律 a = $g_n(W_nX_n+\theta_n)$ ,$X_n = g_{n-1}(W_{n-1} X_{n-1}+\theta_{n-1})$... 

##  多层神经网络的架构
说到架构,我也有点虚,我只能说说我现在的理解.可能不对,如果发现希望指出,我也会随时间有不同的理解吧,所以标注上时间.

(2019/05/29)
架构,算是定义了神经元彼此的连接方式,计算方式和训练策略.

 - 从链接方式上说, 就是描述了神经元要怎么链接在一起. 比如从最简单的单层,到多层(MLP), 链接越来越深, 递归(RNN)把上一次的输出也作为输入等等.
 - 从计算方式上说, 就是描述了如何对输入进行处理. 比如卷积(CNN)就是引入了卷积操作.bathNormal, dropout诸如此类的.
 - 从训练策略上说, 就是描述了如何训练神经网络,让结果越来越好. 比如对抗生成网络(GAN) 强化学习(RL)
