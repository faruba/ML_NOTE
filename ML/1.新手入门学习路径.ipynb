# 学习路径

## 概要
  这一章主要内容是为像我这样的初学者，提供一个可行的学习路径（从0到入门）
 - 先看 《Python 神经网络编程》 作者 塔里克。拉希德。 推荐原因：不需要数学功底，Python编程经验。从0到有把神经网络的知识讲的外焦里嫩。
 - 再看 吴恩达的课程 网易云课堂免费提供 5门课 优势是免费，缺点是没有习题。不过GitHub上有很多人这门课的作业。可以fork 作为练习（建议先删答案，自己做一遍，然后按照自己理解重新实现一遍）
 - 如果数学不好，而且感觉数学已经成为阻碍学习的障碍（基本的概率，线代，微积分知识还是要有的）那么数学书我建议《程序员的数学》 概率和线代 作者平冈和幸。
 - 我觉得**英文**很重要。但我英文又不好，我的方法是，如果书籍有英汉的，我会两个版本都下载下来，对照着看，这样表达方式，不会的单词就慢慢的都会了。视频如果是英文的，我会尽量找双语字幕，先不开字幕，尽量听， 然后开英语字幕再听，最后开中文字幕，虽然学习慢了很多。但我个人觉得这样的付出是值得的。 不认识的单词坚持查字典,并把单词记录到**百词斩** 

好。以上就是入门前的建议。如果有兴趣，可以继续看。后面主要记录的是自己的学习历程和感悟。

## 初学经历

初次接触AI。我在网上搜索了很多资料。目的是为了给自己制定一个可行的学习计划。面对大量的信息，意识到AI的坑非常深。
**根据以往的经验，想爬陡峭的山，就不要看山的整体，免得失去爬山的信心，just do it**

  我在知乎上搜索类似“新手如何自学神经网络”的提问。按照他们的建议下载了相关书籍，计划先看一看，看看哪里不会，再想办法补足。 他们推荐的是**花书**(Deep Learning: Adaptive Computation and Machine Learning series)。前几章都很简单，更确切的说很符合直觉。然后突然间海量的公式出现在眼前。一脸懵逼。一是看不懂公式的含义（数学功底差），二是不理解为什么要这么做（他们从不讲背后的思维过程） 在此我有些犹豫，想停下来先恶补一下数学，又担心学习数学上的挫折会让我放弃（数学我已经放弃了很多次了） 既然数学不是短时间就可以补足的，也就只能硬着头皮看。考虑到数学这块硬骨头暂时啃不动，**就改变期望：尽量理解**，从比较宏观的视角看书，忽略很多不懂的技术细节。比如我大体知道各个技术是为了解决什么问题，优缺点是什么。但为什么会有这些问题，为什么可以解决这些问题，则完全不懂。虽然一知半解，但好在建立起了一个**笼统的认识**。这相当于一个地图，避免自己迷路。

  *反思一下，数学算是我遇到的第一个坑。 数学是否重要，我下面会单独说一下我的感受。我先说说我对数学对影响我学习AI的态度：
  
  1. 这是一个好事：数学之所以会被发明，其实就是人遇到了困难然后解决后得到的胜利果实。既然学习过程中遇到了看不懂的数学问题，就说明我没看清楚这个数学在试图解决什么问题。想明白这个，比看懂这个数学知识更重要。
  2. 这是一个很好了解智能的机会，记录过程，分析过程，会得到很多感悟。会发现很多AI上的技术在学习过程中的体现，比如决策森林，LSTM，注意力机制。

为了解决数学的短板。我买了2本数学书《程序员的数学》 概率和线代 作者平冈和幸。买这本书原因有二:
 1. 我是程序猿，这本书比较符合程序猿思维 
 2. 本书不要求有很高的数学功底（小学水平？坐标系什么时候学的来着？）。
读后感：爽！。这本书不像大学用的那种教科书，充斥着各种晦涩难懂的定义和推理。他把矩阵，概率诸多性质全部剔除，观察分析他们的样子性质和能力极限，然后一点一点把剔除的性质一点一点加回来。让我从最开始就得以明白他们的本质。 这本书让我受用最大的，应该是让我不再那么惧怕数学了，数学这东西，就应该像学历史那样学。**重要的是数学思维,而非数学本身**


 *反思一下，数学是否真的那么重要。其实挺重要的，但是作为初学者，数学不那么重要。我遇到的第一个坑是数学，其实现在看来其实应该是选错了书，直接看我推荐的Python神经网络就不会有这种痛苦了，不过塞翁失马焉知非福，没遇到数学坑也就没机会遇到这本书了吧。之前看有人在知乎上说搞AI，大体3个境界，第一境界是用人家写好的代码，喂喂数据，更进一步按照里面的说明做迁移学习，就可以得到自己想要的结果。难度很低。第二境界是调优，这里会涉及到改架构，调超参，数据处理，结果分析等诸多技术。这一境界挺宽的，我想有数学功底肯定会有莫大的帮助。不过据我所知，更多的是利用工程学方法，通过大量的实验结果选择最适合的架构，调优。第三境界就是创建架构了吧。要做到这一步。我想数学必须要很精通。所以如果看到很多公式看不懂，不要怕，很多时候影响不大（我最近在根据一篇论文写代码，最后损失函数那部分因为数学不好，2天都毫无进展，但更主要的原因是我没有学会走就想跑。我知道，是因为一些基础知识（GAN）我还没有掌握才会这样）。考虑到数学的困难程度。还是早点下手比较好。免得到后面举步维艰。


关于推荐学习路径，我个人的看法是这样的：
 1. 要知道自己的学习特点. 有人喜欢看书学习(比如我),有人喜欢看视频教程.根据自己的喜好找材料.
 2. 别人推荐的，不一定是最适合自己的。每个人的知识储备和结构都不一样。可能某些书很适合我，却未必适合在座的你。
 
所以我想，分享如何选适合自己的书才是更加有益的吧。


对于我不了解的领域，我选书的策略是这样的：先听取别人的建议，然后看一看，看能不能看懂。能看懂，说明适合，那就按照这个人的建议去做。如果看不懂，就**分析自己缺少哪些知识**。然后去找相关资料。找到能看懂的资料后，学会并吸收。这些资料如果还推荐了其他书籍，先记录下来。感觉自己有信心看懂上一个看不懂的书了，就试试。如果还看不懂，说明这本书依旧不太适合自己。那就把之前能看懂的那些书推荐的书，拿过来看。如果没有推荐的书，可以查查作者的介绍，还出过什么书，如果是外文书，书的翻译人还翻译过哪些书,这样就可以发掘出许多书。用同样的方法去学习。不断积累。学习就像爬楼，书就像梯子，但是不同的梯子长度不同，起始高度也不同，找合适的书就是让这些梯子可以连在一起。

我描述一下我认为是好书的特点：
 1. 直指本质：这种感觉就像一开始就给一个骨架，简简单单，一眼就可以看透，然后一点一点往上贴泥，每加一点，就增加一些特性。从简单到复杂，变化万千。每一步都不觉得太难，但回头看着自己走过的路，自己都觉得震撼。
 2. 引人思考：循循善诱，促使人思考其本质，质疑自己的认知，尝试不同的思维方式、
 3. 授人以渔：教会人思考方式，教会人如何提出正确的问题。思考如何运用于实践。


花书我看了3遍花书，还是进展不大，看了我前面介绍的书和视频，现在回头看花书还是很多不明白的。这本书真的是入门书籍吗？

我现在书视频已经看了不少，而且练习也做了。但是很多其他的书依旧看不明白。那现在我要怎么办呢？ 我的计划是这样的：
1. Python 神经网络那本书。不看源码自己写一份。跑一跑数据，看看效果如何。
2. 吴恩达中的期末练习。按照自己的理解重新实现一份。
3. 同类型的实现，在GitHub上搜索相关项目，分析他们的实现。学习现在流行的开源框架。
